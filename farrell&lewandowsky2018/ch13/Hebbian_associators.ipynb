```r
stim <- list( c(1,-1,1,-1), c(1,1,1,1) )
resp <- list( c(1,1,-1,-1), c(1,-1,-1,1) )

n <-4 # number of input units
m <-4 # number of output unit s


W <- matrix ( rep (0 ,m*n) , nrow=m)
print(W)
alpha <- 0.25
```

```r
# Learning
for (pair in 1:2) { # store association for each pair
    print(W)
    print('stim = ')
    print(stim[[pair]])
    print('resp = ')
    print(resp[[pair]])
    for (i in 1 :m) { # loop across output units
        for (j in 1 :n) { # loop across input units
            W[i , j] <- W[i , j] + alpha * stim[[pair]][ j] * resp[[ pair ]][ i]}
        print(W)
    }
}
```

```r
# Test phase; test with first stimulus
o <- rep (0 ,m)
for (i in 1 :m) {
    for (j in 1 :n) {
        print(o)
        o[i] <- o[i] + W[i , j] * stim [[1]][j]
    }
}

require(lsa)
print(stim[1])
print(o)
cosine(o , resp [[1]])
```

```r
library (lsa)
n <- 100 # number of input unit s
m <- 50 # number of output unit s

listLength <- 20 # number of pairs in each list

nReps <- 100

alpha <- .25

stimSimSet <- c (0 , .25 , .5 , .75 , 1)

accuracy <- rep (0 , length (stimSimSet) )

for ( rep in 1 :1) {
    #initialize weights
    W <- matrix(rep(0,m*n),nrow=m)

    stim1 <- {}
    resp1 <- {}

    # create study set
    for (litem in 1 :listLength) {
        #random list of numbers
        svec <- sign ( rnorm (n) )
        stim1 <- c (stim1 , list(svec))
        
        rvec <- sign ( rnorm (m) )
        resp1 <- c (resp1 , list(rvec))
    }

    # study list
    #for each pair in the set...
    print(W[0:25,0:13])
    for (litem in 1 :listLength) {
        c <- stim1[[litem]]
        o <- resp1[[litem]]
        #this is where the actual "learning" occurs
        #the weights are set as the dot product of the stimulus and the response.T
        W <- W + alpha * o%*%t(c)
    }
    print(W[0:25,0:13])
    # loop across probe stimuli of differing similarity to
    # the trained stimuli , and use these to probe for
    # responses
    for (stimSimI in 1 : length (stimSimSet) ) {

        stimSim <- stimSimSet[stimSimI]

        # create test stimuli
        stim2 <- {}
        for (litem in 1 :listLength) {
            
            #creates a new stimulus, and the higher the "similarity" parameter is,
            #the more values of the new stimulus will be replaced by the actual training stimulus
            svec <- sign ( rnorm (n) )
            mask <- runif (n)<stimSim
            stim2 <- c (stim2 , list (mask*stim1[[litem]] + (1-mask) * svec) )
        }

        # test list
        tAcc <- 0
        for (litem in 1 :listLength) {
            c <- stim2[[litem]]
            o <- W%*%c
            tAcc <- tAcc + cosine(as.vector(o), resp1[[litem]])
        }

        accuracy[stimSimI] <- accuracy[stimSimI] + tAcc / listLength

    }
} # end reps loop

# pdf('HebbGraceful.pdf', width=5, height=5)
plot(stimSimSet , accuracy / nReps , type='b', xlab='Stimulus−Cue Similarity', ylab='Cosine' )
dev.off()
```

```r
library (lsa)

n <- 100 # number of input unit s
m <- 50 # number of output unit s

listLength <- 20 # number of pai r s in each li st

nReps <- 100

alpha <- .25

lesionPSet <- seq (0 ,1 , by = 0.1)

accuracy <- rep (0 , length (lesionPSet) )

for ( rep in 1 :nReps) {
#W <− mat rix ( rnorm (m∗n , s d = . 1 ) , nrow=m)
W <- matrix(rep(0 , m*n) , nrow=m)

stim1 <- {}
resp1 <- {}

# create study set
for (litem in 1 :listLength) {

    svec <- sign ( rnorm (n) )
    stim1 <- c (stim1 , list (svec) )

    rvec <- sign ( rnorm (m) )
    resp1 <- c (resp1 , list (rvec) )

}

# study list
for (litem in 1 :listLength) {
    c <- stim1[[litem]]
    o <- resp1[[litem]]
    W <- W + alpha*o%*%t ( c )
    #W <− W + alpha ∗ tcrossprod (o,c)
}

for (lesionPI in 1 : length (lesionPSet) ) {

lesionP <- lesionPSet[lesionPI]
Wlesion <- W
mask <- matrix ( runif (m*n)<lesionP , nrow=m)
Wlesion[mask]=0

# test list
tAcc <- 0
for (litem in 1 :listLength) {
c <- stim1[[litem]]
o <- Wlesion %*% c

tAcc <- tAcc + cosine(as.vector(o), resp1[[litem]])
}
accuracy[lesionPI] <- accuracy[lesionPI] + tAcc / listLength

}
} # end reps loop

#pdf('HebbLesion.pdf', width=5, height=5)
plot(lesionPSet , accuracy / nReps , type='b', xlab='Lesion Probability', ylab='Cosine', ylim=c(0 ,1))
dev.off()
```
